```python
import matplotlib.pyplot as plt
```

```python
file = open("initWave.txt", "r")
tmp = []

xarray = []
yarray = []

for line in file:
    tmp.append(float(line))

for id, val in enumerate(tmp):
    if id%2 == 0:
        xarray.append(val)
    else:
        yarray.append(val)
```

```python
plt.scatter(xarray, yarray)
```
